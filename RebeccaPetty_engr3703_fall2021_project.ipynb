## Code for Underdamped System

```python
%matplotlib inline
from math import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

gamma = 0.7 ##for underdamped
k = 48.
##Constants
c1 = 1./6
c2 = 2./6
c3 = 2./6
c4 = 1./6
a2 = 1./2
a3 = 1./2
a4 = 1.
b21 = 1./2
b31 = 0.
b32 = 1./2
b41 = 0.
b42 = 0.
b43 = 1.


F_over_m = lambda t: 0
fxy_x = lambda t, x, v: v
fxy_v = lambda t, x, v: F_over_m(t)-2*gamma*v-k*x

def rk4_x(ti, xi, vi, dt):
    K1 = fxy_x(ti,xi, vi)
    K2 = fxy_x(ti+a2*dt,xi+b21*K1*dt,vi)
    K3 = fxy_x(ti+a3*dt,xi+b31*K1*dt+b32*K2*dt,vi)
    K4 = fxy_x(ti+a4*dt,xi+b41*K1*dt+b42*K2*dt+b43*K3*dt,vi) 
    xip1 = xi+(c1*K1+c2*K2+c3+K3*c4+K4)*dt
    return xip1

def rk4_v(ti, xi, vi, dt):
    K1 = fxy_v(ti, xi, vi)
    K2 = fxy_v(ti+a2*dt,xi,vi+b21*K1*dt)
    K3 = fxy_v(ti+a3*dt,xi,vi+b31*K1*dt+b32*K2*dt)
    K4 = fxy_v(ti+a4*dt,xi,vi+b41*K1*dt+b42*K2*dt+b43*K3*dt)
    vip1 = vi+(c1*K1+c2*K2+c3*K3+c4*K4)*dt
    return vip1

ti = 0.0
tf = 5.0
n = 5000
dt = (tf-ti)/n
t = np.zeros(n+1)
x = np.zeros(n+1)
v = np.zeros(n+1)
x_iv = 1.0
v_iv = -0.4
t[0] = ti
x[0] = x_iv
v[0] = v_iv


for i in range(1, n+1):
    t[i] = ti + i*dt
    x[i] = rk4_x(t[i-1],x[i-1],v[i-1],dt)
    v[i] = rk4_v(t[i-1],x[i-1],v[i-1],dt)
    
##fig = plt.figure(figsize=(6,6))
plt.plot(t,x,label="x vs. t",color="b",linewidth="2.0")
plt.title("Displacement for an Underdamped System")
plt.tight_layout()
plt.show()

##fig = plt.figure(figsize=(6,6))
plt.plot(t, x, label = "v vs. t", color = "b", linewidth = "2.0")
plt.title("Velocity for an Underdamped System")
plt.legend()
plt.tight_layout()
plt.show()



```
